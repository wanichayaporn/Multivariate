#Transpose

```
A = np.array([[3,-1,2],[1,5,4]])
print(A)
```

```
A.T
```

#Matrix Multiply

$B =  A*A' $

```
B = np.dot(A,A.T)
print(B)
```

#Determinant

$Det(B)$

```
np.linalg.det(B)
```
